```python
#importing the california house price data
from sklearn.datasets import fetch_california_housing
```

```python
dataset = fetch_california_housing()
```

```python
type(dataset)
```

```python
#the dataset is in dictionary and we have two key which is data and target.
print(dataset)
```

```python
#pandas dataframe
#here we only include data from dataset and not target. for the data which contain different columns we use feature names as column names
data = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```
