# **Label Encoding**
converting the text labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_data = pd.read_csv('breast_cancer_data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

here
*   0 --> benign
*   1 --> malignant
*  here 0 is given to benign because of alphabetical order


```python
cancer_data['target'].value_counts()
```

```python
#as we created a column name target with the data of diagnosis column, we are going to remove diagnosis column
cancer_data = cancer_data.drop('diagnosis', axis=1)
```

```python
cancer_data.head()
```
