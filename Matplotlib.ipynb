#Matplotlib
**useful for making plots and so used for data visualization**

```python
#importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100) #it will give a equal space value between 0 to 10.
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data

#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x axis,y axis labels
plt.plot(x,y)
plt.xlabel('angel')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola

x = np.linspace(-10,10,20)
y = x*x
plt.plot(x,y)
plt.show()
```

```python
#here r means red and + is for giving + symbol.
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), 'g-')
plt.plot(x,np.cos(x), 'r--')
```

```python
#bar plot

fig = plt.figure() #empty plot
ax = fig.add_axes([0,0,1,1])  #left,bottom, height,width - for rectangle
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100,50,10,40,70]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

```python
#pie chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100,50,10,40,70]
ax.pie(people,labels = languages, autopct = '%1.1f%%') # autopct for decimal points
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'b')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap = 'Blues')
plt.show()
```
